```python
import numpy as np
```

## Numpy basics and numpy array

```python
# creating array from list
arr_1d = np.array([1, 2, 3, 4, 5])
print("1D array: ", arr_1d)
print()
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D array: ", arr_2d)
print()


# creating array from scratch
zeros = np.zeros((3, 4))
print("zeros array: \n", zeros)
print()

ones = np.ones((2, 3))
print("one array: \n", ones)
print()

full = np.full((2, 2), 7)
print("full array: \n", full)
print()

random = np.random.random((2, 3))
print("random array: \n", random)
print()

sequence = np.arange(0, 11, 2)
print("sequnce array: \n", sequence)

identity = np.eye(4)
print("\nidentity array: \n", identity)
```

## Vector, Matrices and Tensors

```python

vector = np.array([1, 2, 3])
print("Vector: ", vector)

matrix = np.array([[1, 2, 3], 
                   [4, 5, 6]])
print("\nMatrix: ", matrix)

tensor = np.array([[[1, 2], [3, 4]], 
                   [[5, 6], [7, 8]]])
print("\nTensor: ", tensor)

arr = np.array([[1, 2, 3], 
                [4, 5, 6]])
print("\nShape :", arr.shape)
print("Dimension :", arr.ndim)
print("Size :", arr.size)
print("Data Type :", arr.dtype)
print("\nfloat values : \n", arr.astype(float)) #changing datatype of array
print()
print()

```

## Properties of Numpy array

```python
# Array Reshaping
arr = np.arange(12)
print("Original array ", arr)

reshaped = arr.reshape((3, 4))
print("\n Reshaped array ", reshaped)

flattened = reshaped.flatten()
print("\n Flattened array ", flattened)

# ravel (returns view, instead of copy)
raveled = reshaped.ravel()
print("\n raveled array ", raveled)

# Transpose
transpose = reshaped.T
print("\n Transposed array ", transpose)
```
