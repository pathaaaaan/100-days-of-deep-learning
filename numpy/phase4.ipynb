```python
import numpy as np
```

### Advance Functions and Manipulation

```python
arr = np.arange(10, 81,10)
print(f"original : {arr}")

new_arr = np.insert(arr, 2, 100, None)
print(f"new array : {new_arr}")
     # np.insert(arr, index, value, axis=None)
```

```python
arr2d = np.array([[1,2],[3,4]])
newarr = np.insert(arr2d, 1, [5,6], axis=0)
print(f"new array : \n{newarr}")
```

```python
# append
arr = np.array([10,20,30])

newarr = np.append(arr,[40,50,60])
newarr
```

```python
# concatenate
arr1 = np.array([10,20,30])
arr2 = np.array([40,50,60])
newarr = np.concatenate((arr1,arr2))
newarr
```

```python
# removing elements
arr = np.array([10,20,30,40,50,60])
```
