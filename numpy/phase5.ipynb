```python
import numpy as np

prices = np.arange(100, 1001, 100)
print(f"prices : {prices}\n")
```

```python
# 10% discount
disc = 10
amt = prices - (prices * disc / 100)
amt
```

```python
#  double prices
double_prices = prices * 2
print(f"double_prices : {double_prices}\n")
```

```python
# 
matrix = np.array([[1,2,3,], [4,5,6]])  # 2 x 3
vector = np.array([7,8,9])      # 1D array with 3 elements

result = matrix + vector    # vec gets added to each rpw pf the matrix
result
```

## Vectorization

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

res = arr1 + arr2   # elememt-wise addition
res
```

```python
mult = arr1 * arr2  # elememt-wise multiplication
mult
```

```python
#  logical operations
a1 = np.array([10, 20, 30])
res = a1 > 15
print(res)

```

```python
a1= np.array([[1, 2], [3, 4]])
a2 = np.array([[5, 6], [7, 8]])
res = np.dot(a1, a2) 
print(res)
```
