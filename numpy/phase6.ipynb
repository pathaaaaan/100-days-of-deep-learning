## handling missing values

```python
import numpy as np
```

```python
arr = np.array([1,2,4,np.nan,5])
print(np.isnan(arr))
```

```python
# replacing a nan value to number i.e. 0
print(np.nan_to_num(arr))
```

```python
arr = np.array([1,2,3,np.inf, 5])
print(np.isinf(arr))
```

```python
cleaned = np.nan_to_num(arr, posinf = 1000)
cleaned
```
