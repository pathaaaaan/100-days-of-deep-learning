## advance operation with business examples

```python
import numpy as np

# [restr_id,2021,2022,2023,2024]
data = np.array([
    [1,150000,180000,220000,250000],     #Paradise Biryani
    [2,120000,140000,160000,190000],     #Biryani House
    [2,200000,230000,260000,300000],     #pizza hut
    [2,180000,210000,240000,270000],     #chai sutta
    [2,160000,185000,205000,230000],     #dominos
])
```

```python

print("==== Zonato sales analysis =====")

print("data shape", data.shape)
print("data for first 3 restaurants\n", data[:3]) 
```

```python
# total sales per year
total_sales = np.sum(data[:,1:],axis = 0)
total_sales
```

```python
# minimum sales per restaurant
min_sales = np.min(data[:,1:],axis = 1)
min_sales
```

```python
# avg sales per restaurant
avg_sales = np.mean(data[:,1:], axis  = 1)
avg_sales
```

```python
cum_sum = np.cumsum(data[:,1:], axis = 1)
cum_sum
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))  
plt.plot(np.mean(cum_sum, axis=0))
plt.title("Average Cummulitative sum")
plt.xlabel("year")
plt.ylabel("cummulative sum")
plt.show()
```
