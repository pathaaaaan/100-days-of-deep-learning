### WEb Scraping

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
webpage = requests.get('https://www.ambitionbox.com/list-of-companies', headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'html.parser')
```

```python
soup.prettify()
```

```python
soup.find_all('h1')[0].text.strip()
```

```python
for i in soup.find_all('h2'):
    print(i.text.strip())
```

```python
company = soup.find_all('div', class_='companyCardWrapper')
```

```python
len(company)
```

```python
name = []
rating = []
for i in company:
    h2 = i.find('h2')
    h2 = h2.text.strip()
    name.append(h2)
    p = i.find('p', class_='companyCardWrapper__companyRatingCount')
    if p:
        rating.append(p.text.strip())
    else:
        rating.append("No rating available")
```
