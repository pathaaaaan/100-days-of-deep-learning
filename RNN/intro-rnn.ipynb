```python
from keras import Sequential
from keras.layers import SimpleRNN, Dense
```

```python
model = Sequential()
model.add(SimpleRNN(3, input_shape=(4,5)))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
print(f"Number of weights in network from input to hidden: {model.get_weights()[0].shape}")
model.get_weights()[0]
```

```python
print(f"Number of weights in feedback loop: {model.get_weights()[1].shape}")
model.get_weights()[1]
```

```python
print(f"Number of weights in network from  hidden to output: {model.get_weights()[2].shape}")
model.get_weights()[2]
```
