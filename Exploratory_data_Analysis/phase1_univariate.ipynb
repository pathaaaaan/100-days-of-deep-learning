### Type of Data
- Numerical (Height, age, etc)
- Categorical (gender, alive, etc)

```python
import pandas as pd
import kagglehub
import os

# Download latest version
path = kagglehub.dataset_download("yasserh/titanic-dataset")

print("Path to dataset files:", path)
file_name = "Titanic-Dataset.csv"
train_path = os.path.join(path, file_name)

# Load it into pandas
df = pd.read_csv(train_path)
df.head()
```

## 1. Categorical Data

```python
import seaborn as sns
```

### countplot

```python
# count plot
# sns.countplot(x = df['Survived'], data= df, palette='Set2')
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(x = df['Pclass'], data= df, palette='Set1')
```

```python
sns.countplot(x = df['Embarked'], data= df, palette='Set2')
```

### piechart

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

```python
df['Embarked'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

## 2. Numerical Data

```python
# histogram
df['Age'].plot(kind='hist')
```

```python
# displot
sns.displot(df['Age'], kde=True)
```

### BOX PLOT

```python
# boxplot used to identify outliers
sns.boxplot(df['Fare'])
```
